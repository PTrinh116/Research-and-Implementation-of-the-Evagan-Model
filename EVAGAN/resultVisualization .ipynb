```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset2014="/content/ACGAN/2014_ACGAN_CC_LISTS.csv"
dataset2017="/content/ACGAN/CIC2017_ACGAN_CC_LISTS.csv"
dataset2018="/content/ACGAN/CIC2018_ACGAN_CC_LISTS.csv"
df1A=pd.read_csv(dataset2014)
df2A=pd.read_csv(dataset2017)
df3A=pd.read_csv(dataset2018)
```

```python
df2A.head()
```

```python
import matplotlib.pyplot as plt
# Tạo subplot kích thước 1 x 3
fig, ax = plt.subplots(1, 3, figsize=(15, 8))
index = 0
colors=['green','orange', 'blue']

dataframes={"ISCX-2014": df1A,"CIC-2017": df2A,"CIC-2018": df3A}

for title, df in dataframes.items():
  df.plot(x="Unnamed: 0", y=["D_Loss_Real","D_Loss_Fake","G_Loss"],
        kind="line", ax=ax[index], title=title, xlabel="Epochs", color=colors)
  index=index+1

fig.tight_layout(pad=5.0)
fig.supylabel('ACGAN')
plt.show()
```

```python
# Tạo subplot kích thước 1 x 3

fig, ax = plt.subplots(1, 3, figsize=(15, 8))
index = 0
colors=['blue','green','red']

dataframes={"ISCX-2014": df1A,"CIC-2017": df2A,"CIC-2018": df3A}

for title, df in dataframes.items():
  df.plot(x="Unnamed: 0", y=["GEN_VALIDITY","NORMAL_EST","REAL_BOT_EVA"],
        kind="line", ax=ax[index], title=title, xlabel="Epochs", color=colors)
  index=index+1

fig.tight_layout(pad=5.0)
fig.supylabel('ACGAN')
plt.show()
```
